# Part-2: Working with Model


```python
# 1) Create the target data and feature data where target data is price
x=dummy_df.drop('Price',axis=1)
x.head().shape
```

```python
y=dummy_df[['Price']]
y.head().shape
```

```python
# 2) Create a linear regression model for Target and feature data
rm=LinearRegression()
rm
```

```python
# rm.fit(x,y)
```

```python
# r2=rm.score(x,y)
# r2
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=10)
```

```python
rm.fit(x_train,y_train)
```

```python
rm.score(x_train,y_train)
```

```python
rm.score(x_test,y_test)
```

```python
price_predicted=rm.predict(x_test)
x_test.head()
```

```python
r2=r2_score(y_test,price_predicted)
r2
```

```python
mse=mean_squared_error(y_test,price_predicted)
mse
```

```python
mae=mean_absolute_error(y_test,price_predicted)
mae
```

```python
rmse=np.sqrt(mse)
rmse
```

```python
slope=rm.coef_
slope.argmax()
```

```python
intercept=rm.intercept_
intercept
```
